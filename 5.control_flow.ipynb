# Loops

**while Loops**

The
execution of a while loop begins with a test of the Boolean condition. If that condition
evaluates to True, the body of the loop is performed.

```
count = 0
while(count < 3):
  count = count + 1
  print("Hello chandu")
```

```
count = 0
while(count < 3):
  count = count + 1
  print("Hello chandu")
else:                     #combining else with while
  print("How are you?")
```

**For Loops**

The for-loop syntax can be used on any
type of iterable structure, such as a list, tuple str, set, dict, or file.

```
n = 4
for i in range(0, n):
  print(i)
```

```
print("List Iteration")
j = ["sai", "chand", "reddy"]
for i in j:
  print(i)
```

```
print("Tuple Iteration")
o = ('sai', 'chand', 'reddy')
for i in o:
  print(i)
```

```
print("string Iteration")
s = "reddy"
for i in s:
  print(i)
```

```
print("Dictionary Iteration")

h = dict()
h['xyz'] = 123
h['abc'] = 345
for i in h:
  print(i)
```

**Break and Continue Statements**

Python supports a **break** statement that immediately terminate a while or for loop
when executed within its body

Python also supports a **continue** statement that causes the current iteration of a
loop body to stop, but with subsequent passes of the loop proceeding as expected.

```
for letters in 'saichandreddy':
  if letters == 'e' or letters == 's':
    continue

  print('current letter :',letters)
  var = 10 
```

```
for letter in 'saichandreddy':
  if letter == 'a' or letter == 's':
    break

print("current number :",letter)
```
