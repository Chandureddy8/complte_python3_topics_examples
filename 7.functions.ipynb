**Functions**

A function is a set of statements that take inputs, do some specific computation and produces output.

```
data = [1,2,3,4,5,6,3]
target = 4

def count(data,target):
  n = 0
  for item in data:
    if item == target:
      n += 1
  return n
count(data,target)
```

```
def swap(x, y):
  temp = x;
  x = y;
  y = temp;

x = 2
y = 3

swap(x,y)
print(x)
print(y)
```

**Default arguments**

A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.

```
def myfun(x, y = 50):
  print("x: ", x)
  print("y: ", y)

myfun(10)
```

**Keyword arguments**

The idea is to allow caller to specify argument name with values so that caller does not need to remember order of parameters

```
def student(firstname, lastname):  
     print(firstname, lastname)  
    
    
# Keyword arguments                   
student(firstname ='chandu', lastname ='reddy')     
student(lastname ='reddy', firstname ='chandu')
```

```
def myfun(*args):
  for arg in args:  #*args for variable number of arguments
    print(arg)

myfun('hello','welcome','to','my_village')
```

```
def Myfun(arg1, *argv):
  print('First argument :', arg1)
  for arg in argv:
    print('Next argument through *argv :',arg)
Myfun('Hello', 'Welcome','to','myworld')
```

```
def myhappy(**kwargs):
  for key, value in kwargs.items():
    print("%s == %s" %(key, value))
myhappy(first = 'chandu',mid = 'reddy', last = 'gudise')
```

```
def MYfun(arg1, arg2, arg3):
  print('arg1:', arg1)
  print('arg2:', arg2)
  print('arg3:', arg3)

args = ('chand', 'reddy','gudise')
MYfun(*args)

kwargs = {"arg1":"chandu","arg2":"reddy","arg3":"gudise"}
MYfun(**kwargs)
```

```
# Python program to illustrate cube of a number   
# showing difference between def() and lambda().

def cube(y):
  return y*y*y

g = lambda x : x*x*x
print(g(7))

print(cube(5))
```

```
#The lambda function gets more helpful when used inside a function.

def power(n):
  return lambda a : a ** n

base = power(2)

print("now power is set to 2")

print("8 power of 2 = ",base(8))

base = power(5)
print('now power is set to 5')

print('8 power of 5 =', base(8))
```

**filter() and map()**

```
a = [100, 2, 3, 4, 60, 4 , 5, 55, 34, 10]

filtered = filter(lambda x : x % 2 == 0, a)
print(list(filtered))

maped = map(lambda x : x % 2 == 0, a)
print(list(maped))
```

**yield**

```
def simplegenratorfun():
  yield 1
  yield 2
  yield 3

for value in simplegenratorfun():
  print(value)
```

```
# A Python program to generate squares from 1 
# to 100 using yield and therefore generator 
  
# An infinite generator function that prints 
# next square number. It starts with 1

def nextsquare():
  i = 1

  while True:
    yield i * i
    i += 1

for num in nextsquare():
  if num > 100:
    break
  print(num)
```
